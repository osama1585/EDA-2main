## Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

## Ans-

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Google Play Store dataset
google_playstore_df = pd.read_csv('https://raw.githubusercontent.com/enjoykcc456/google-playstore-analysis/main/googleplaystore.csv')

# Drop rows with missing values in the 'Rating' and 'Size' columns
google_playstore_df = google_playstore_df.dropna(subset=['Rating', 'Size'])

# Replace 'Varies with device' with NaN and drop those rows
google_playstore_df['Size'] = google_playstore_df['Size'].replace('Varies with device', pd.NA)
google_playstore_df = google_playstore_df.dropna(subset=['Size'])

# Remove commas from size values and convert to numeric
google_playstore_df['Size'] = google_playstore_df['Size'].str.replace(',', '').str.extract(r'(\d+\.*\d*)').astype(float)

# Create a scatter plot to visualize the relationship between app size and rating
plt.figure(figsize=(10, 6))
plt.scatter(google_playstore_df['Size'], google_playstore_df['Rating'], alpha=0.5)
plt.xlabel('App Size (MB)')
plt.ylabel('Rating')
plt.title('Relationship between App Size and Rating')
plt.grid(True)
plt.show()

```

## Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

## Ans-

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Google Play Store dataset
google_playstore_df = pd.read_csv('https://raw.githubusercontent.com/enjoykcc456/google-playstore-analysis/main/googleplaystore.csv')

# Drop rows with missing values in the 'Type' and 'Price' columns
google_playstore_df = google_playstore_df.dropna(subset=['Type', 'Price'])

# Remove rows where the 'Price' column contains non-numeric values
google_playstore_df = google_playstore_df[google_playstore_df['Price'].str.replace('.', '', 1).str.isdigit()]

# Remove the dollar sign from the 'Price' column and convert it to numeric
google_playstore_df['Price'] = google_playstore_df['Price'].str.replace('$', '').astype(float)

# Calculate average prices by app type
average_prices = google_playstore_df.groupby('Type')['Price'].mean()

# Create a bar chart to compare average prices by app type
plt.figure(figsize=(10, 6))
average_prices.plot(kind='bar', color=['skyblue', 'salmon'])
plt.xlabel('App Type')
plt.ylabel('Average Price ($)')
plt.title('Average Prices by App Type')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

```

## Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

# Ans-

```python
import pandas as pd

# Load the Google Play Store dataset
google_playstore_df = pd.read_csv('https://raw.githubusercontent.com/enjoykcc456/google-playstore-analysis/main/googleplaystore.csv')

# Drop rows with missing values in the 'Installs' column
google_playstore_df = google_playstore_df.dropna(subset=['Installs'])

# Remove non-numeric characters from the 'Installs' column
google_playstore_df['Installs'] = google_playstore_df['Installs'].str.replace('[^0-9]', '', regex=True)

# Convert the 'Installs' column to numeric
google_playstore_df['Installs'] = pd.to_numeric(google_playstore_df['Installs'], errors='coerce')

# Drop rows with NaN values in the 'Installs' column
google_playstore_df = google_playstore_df.dropna(subset=['Installs'])

# Group the dataset by app and calculate the total number of installs for each app
app_installs = google_playstore_df.groupby('App')['Installs'].sum()

# Sort the app installs in descending order and select the top 10
top_10_apps = app_installs.sort_values(ascending=False).head(10)

# Print the top 10 most popular apps
print("Top 10 Most Popular Apps:")
print(top_10_apps)
```
